```python
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
bit = pd.read_csv('./BitCoin.csv')

bit
```

```python
bit_period = bit[(bit.index >= 84) & (bit.index <= 478)]

bit_period[(bit_period.index == 84) | (bit_period.index == 478)]
```

```python
def fix_date(x):
    if (x[5:7] >= "01") & (x[5:7] <= "12") & (x[8:10] < "13"):
        return x[0:4] + "-" + x[8:10] + "-" + x[5:7]
    else :
        return x

bit_period["Fixed_Date"] = bit_period["Date"].map(fix_date)
bit_period["Fixed_Date_Transformed"] = pd.to_datetime(bit_period["Date"].map(fix_date), format="%Y-%m-%d")
bit_period.head(40)
```

```python
bitcoin = pd.read_csv('BitCoin.csv')
t0 = bitcoin[bitcoin['Date'] == '2016-06-01'].index[0]
t1 = bitcoin[bitcoin['Date'] == '2017-06-30'].index[0]
ma5 = bitcoin['Open'][t0:t1-1:-1].rolling(5, min_periods=1, center=True).mean()
plt.figure(figsize=(10,6))
plt.title('BitCoin Price')
plt.plot(bitcoin['Date'][t0:t1-1:-1],ma5,c='#f2a900')
plt.ylabel('$')
plt.xlabel('Date')
plt.xticks(bitcoin['Date'][t0:t1-1:-90])
plt.show()
```
