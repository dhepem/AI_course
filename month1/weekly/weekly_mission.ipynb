넷플릭스에서 가장 많은 작품을 만든 나라가 어디이고, 얼마나 많은 작품을 produce했는지 알아보자.

가장 많은 작품이 올라간 국가는 어디이고, 얼마나 많은 작품이 있는가?  
country column을 기준으로 한다.  
단일 국가인 경우를 기준으로 결과를 구해보자.

```python
net_single_country = net.dropna(axis=0, subset=['country'])

net_single_country.info()
```

```python
net_single_country = net_single_country[~(net_single_country['country'].str.contains(','))]

net_single_country
```

```python
net_single_country_cnt = net_single_country['show_id'].groupby(by=net_single_country['country']).count()

net_single_country_cnt
```

```python
net_single_country_cnt[net_single_country_cnt == max(net_single_country_cnt)]
```

2017년 한 해의 가격의 변화를 보여주려고 한다. 그런데 매일매일의 데이터를 보여주는 것보다 이것의 추세를 표현해주면 좋겠다는 생각이 들어 Moving Average(이동평균법) 를 도입하고자 한다.

2016.6 ~ 2017.6 기간의 5-MA(Moving Average) 비트코인 가격 그래프를 그려주세요.

선의 색깔은 노란색(#f2a900) 으로 해야합니다.
x-axis label과 y-axis label은 적절한 텍스트로 추가되어야 합니다.
이동평균의 기준은 Open Column으로 계산합니다.
이외에도 그래프를 알아보기 쉽게 하기 위한 정보(Text, Facet 등)을 추가하는 것도 좋습니다.
💡 이동평균(Moving Average)법은 시계열 데이터를 표현하는 데에 많이 사용하는 방법 중 하나입니다.

```python
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
bit = pd.read_csv('./BitCoin.csv')

bit
```

```python
bit_period = bit[(bit.index >= 84) & (bit.index <= 478)]

bit_period[(bit_period.index == 84) | (bit_period.index == 478)]
```

```python
def fix_date(x):
    if (x[5:7] >= "01") & (x[5:7] <= "12") & (x[8:10] < "13"):
        return x[0:4] + "-" + x[8:10] + "-" + x[5:7]
    else :
        return x

bit_period["Fixed_Date"] = bit_period["Date"].map(fix_date)
bit_period["Fixed_Date_Transformed"] = pd.to_datetime(bit_period["Date"].map(fix_date), format="%Y-%m-%d")
bit_period.head(40)
```

```python
bitcoin = pd.read_csv('BitCoin.csv')
t0 = bitcoin[bitcoin['Date'] == '2016-06-01'].index[0]
t1 = bitcoin[bitcoin['Date'] == '2017-06-30'].index[0]
ma5 = bitcoin['Open'][t0:t1-1:-1].rolling(5, min_periods=1, center=True).mean()
plt.figure(figsize=(10,6))
plt.title('BitCoin Price')
plt.plot(bitcoin['Date'][t0:t1-1:-1],ma5,c='#f2a900')
plt.ylabel('$')
plt.xlabel('Date')
plt.xticks(bitcoin['Date'][t0:t1-1:-90])
plt.show()
```
